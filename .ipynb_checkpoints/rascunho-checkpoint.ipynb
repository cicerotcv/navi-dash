```python
import pandas as pd
from string import ascii_letters
from random import randint
from json import dumps
from os import listdir
```

```python
# REACT_DATA_DIR = "./navi-dash/src/data/" # utilizado no desenvolvimento
REACT_DATA_DIR = "./data/"
INPUT_DIR = "./data/"
```

```python
def get_id():
    return hex(int(str(randint(1000, 100000)), 36)).lstrip('0x')

def save_json(filename, dict):
    with open(f"{REACT_DATA_DIR}{filename}", 'w+', encoding='utf-8') as file:
        file.write(dumps(dict))

def id_from_sectorname(name):
    for sector in sectors:
        if sector['sector'] == name:
            return sector['id']
```

```python
df = pd.read_json(f'{INPUT_DIR}esgepoluentes.json')
```

```python
df
```

```python
sectors = list(df['sector_name'].unique())
sectors = [{"id": get_id(), "sector": sector} for sector in sectors] 
print(len(sectors))
print(len({sector["id"] for sector in sectors}))
save_json("sectors.json", sectors)
```

```python
score_cost = pd.read_json(f'{INPUT_DIR}ESG_gasto.json')
score_cost['sector_id'] = [ id_from_sectorname(name) for name in [*score_cost.index] ]
score_cost = score_cost.reset_index().rename(columns={
    "index":"sector_name",  "Absolute: Air Pollutants Direct & Indirect Cost": "cost"
    })
score_cost
```

```python
score_income = pd.read_json(f'{INPUT_DIR}score_income.json')
score_income['sector_id'] = [ id_from_sectorname(name) for name in [*score_income.index] ]
score_income = score_income.reset_index().rename(
        columns={"index":"sector_name", "Absolute: Air Pollutants Direct & Indirect Cost": "cost", "Valor de Score": 'score'})
score_income
```

```python
score_income_dict = score_income.drop(["sector_name"], axis=1).to_dict()

score_income_dict = [
    {"sector_id": sector_id, "score": score, "cost": cost}
    for (sector_id, score, cost) in zip(score_income_dict['sector_id'].values(), score_income_dict['score'].values(), score_income_dict['cost'].values())
]
```

```python
score_income_dict
```

```python
save_json("score_cost.json", score_income_dict)
```

```python
scatter_data = pd.read_json(f"{INPUT_DIR}scatter.json")
scatter_data = scatter_data.sample(n=250)
```

```python
scatter_data = scatter_data.rename(columns={"Environmental Dimension Score": "score_env", "S&P Global ESG Score":"score_esg"})
scatter_data = scatter_data.to_dict()
petrol_dict = [{"score_env": score_env, "score_esg": score_esg} for score_env, score_esg in zip(scatter_data["score_env"].values(), scatter_data["score_esg"].values())]
```

```python
save_json("petrol_scatter.json", petrol_dict)
```

```python
mercado = pd.read_json(f'{INPUT_DIR}GraficoPLTotaleIndiceIEEB3Comparado.json')
mercado
```

```python
fin_dict = mercado[["IEE B3", "PL Total"]].reset_index().to_numpy()
fin_dict = [{"date": str(date).split(" ")[0], "b3": b3, "pl_total": pl_total} for date, b3, pl_total in fin_dict]
save_json("financial.json", fin_dict)
```
